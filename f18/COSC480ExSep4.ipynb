```python
#Exercise 1 - create a function that takes in 3 parameters: a b c. If a is
#a 0, return b+c, if a is a 1, return b-c, if a is neither, return b * c
def bunchofmath(a, b, c):
    if a == 0:
        return b + c
    elif a == 1:
        return b - c
    else:
        return b * c

print(bunchofmath(0, 4, 5))
print(bunchofmath(2, 4, 5))
```

```python
# Exercise 2 - create a function called facts that passes a
# single parameter x and computes the factorial of x and returns it
# Then ask the user for input, get a number, then call facts(x) and print
# the result.

def facts(x):
    y = x
    for j in range(1, x):
        x = x * (y-j)
    return x

print(facts(int(input("Enter a number: "))))
```

```python
# exercise 3 - create a function that passes in two variables x and y, and 
# returns a list of x random integers from 0 to y, inclusive.
import random

def randList(x, y):
    a = []
    for i in range(x):
        a.append(random.randint(0, y))
    return a

print(randList(5, 10))
```
